### Creating Correlation Coefficient Heat Map and Triangle Correlation Coefficient Heat Map via Python - YOUTUBE CHANNEL(JS_DATA)
#### TASK
##### Use python to display the correlation coefficient btn any two numeric variables in the wine dataset.
##### The value of correlation coefficient would be from -1 to 1


```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv(r'C:\Users\Cru_3d\Desktop\Data Engineering\Python\Projects\Correlation coefficient\Data\winequality.csv')
df.head()
```

```python
df['quality'].value_counts()
```

```python
# Extract all the numerical variables
numerical_df = df.iloc[:, 0:11]
numerical_df.head()
```

```python
corr = numerical_df.corr()
corr
```

```python
# heatmap
plt.figure(figsize=(8, 8))
sns.heatmap(corr, square= True, cmap='bwr', vmin= -1, vmax= 1, linewidths= 0.5)
plt.show()
```

```python
matrix = np.triu(corr)
matrix
```

```python
plt.figure(figsize=(8,8))
sns.heatmap(corr, mask= matrix, square= True, cmap= 'bwr', vmax= 1, vmin= -1, linewidths= 0.8, annot= True)
```
